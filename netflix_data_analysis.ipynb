# Netflix Data Analysis
Exploratory Data Analysis of Netflix Movies and TV Shows dataset using Python, Pandas, Matplotlib, and Seaborn.


## Importing Required Libraries
In this step, we import the Python libraries required for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset
The dataset is loaded using Pandas to begin the analysis.

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

## Understanding the Dataset
This step helps us understand the structure of the dataset, including columns, data types, and missing values.

```python
df.info()
```

## Data Cleaning
Handling missing values in important columns to ensure better analysis.

```python
df = df.dropna(subset=["country"])
```

## Movies vs TV Shows
Analyzing the distribution of Movies and TV Shows available on Netflix.

```python
sns.countplot(x="type", data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

## Top Countries Producing Netflix Content
This visualization shows which countries produce the most content available on Netflix.

```python
df["release_year"].value_counts().sort_index().plot()
plt.title("Content Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries Producing Netflix Content")
plt.show()
```

## Genre Analysis
Identifying the most common genres available on Netflix.

```python
df["listed_in"].value_counts().head(10).plot(kind="bar")
plt.title("Top Genres on Netflix")
plt.show()
```

• Netflix has significantly more Movies than TV Shows.
